```python
import os


directory_name = 'rotate'
try:
    file_list = os.listdir(directory_name + '.flt')
except:
    file_list = os.listdir(directory_name)

new_list_name = directory_name + '_epochs.txt'

epoch = open(new_list_name, "w+")
header = '# rootname   pid vis exp filter epoch     mjd  exptime  darcsec orient\n'
epoch.write(header)


epoch_num = 1
general_string = 'akarai_match_Y106_'
mjd_init = 50200.4

for month in range(11,14):
    
    temp_epoch_list = []
    
    for day in range(1,10):
        for file in file_list:
            file_name = file.split('.')[0]
            fmonth = int(file_name.split('_')[3])
            fday = int(file_name.split('_')[4])
            if fmonth==month:
                if fday==day:
                    temp_epoch_list.append(fday)
                    
    temp_epoch_list.sort()
    day_check = 1
    for placehold in temp_epoch_list:
        if day_check == placehold:
            file_name = general_string + str(month) + '_' + str(placehold) + '  '
            filter1 = 'f105w  ' #this is the hubble equivalent of the Y106 filter
            
            # The following are dummy variables
            pid = '99999  '
            mjd = str(mjd_init) + '  '
            exptime = '600.00  '
            darcsec = '-1.00  '
            orient = '-1.00\n'
            if epoch_num < 10:
                vis = '0' + str(month) + '  '
                exp = '0' + str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + '0' + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
                day_check +=1
            else:
                vis = str(month) + '  '
                exp = str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
                day_check +=1
        
        else:
            epoch_num +=1
            day_check = placehold + 1
            epoch.write('\n')
            
            file_name = general_string + str(month) + '_' + str(placehold) + '  '
            filter1 = 'f105w  ' #this is the hubble equivalent of the Y106 filter
            
            # The following are dummy variables
            pid = '99999  '
            mjd_init += 30
            mjd = str(mjd_init) + '  '
            exptime = '600.00  '
            darcsec = '-1.00  '
            orient = '-1.00\n'
            if epoch_num < 10:
                vis = '0' + str(month) + '  '
                exp = '0' + str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + '0' + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
            else:
                vis = str(month) + '  '
                exp = str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
    
    epoch_num +=1
    epoch.write('\n')
    
    for day in range(10,20):
        for file in file_list:
            file_name = file.split('.')[0]
            fmonth = int(file_name.split('_')[3])
            fday = int(file_name.split('_')[4])
            if fmonth==month:
                if fday==day:
                    temp_epoch_list.append(fday)
                    
    temp_epoch_list.sort()
    day_check = 1
    for placehold in temp_epoch_list:
        if day_check == placehold:
            file_name = general_string + str(month) + '_' + str(placehold) + '  '
            filter1 = 'f105w  ' #this is the hubble equivalent of the Y106 filter
            
            # The following are dummy variables
            pid = '99999  '
            mjd = str(mjd_init) + '  '
            exptime = '600.00  '
            darcsec = '-1.00  '
            orient = '-1.00\n'
            if epoch_num < 10:
                vis = '0' + str(month) + '  '
                exp = '0' + str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + '0' + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
                day_check +=1
            else:
                vis = str(month) + '  '
                exp = str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
                day_check +=1
        
        else:
            epoch_num +=1
            day_check = placehold + 1
            epoch.write('\n')
            
            file_name = general_string + str(month) + '_' + str(placehold) + '  '
            filter1 = 'f105w  ' #this is the hubble equivalent of the Y106 filter
            
            # The following are dummy variables
            pid = '99999  '
            mjd_init += 30
            mjd = str(mjd_init) + '  '
            exptime = '600.00  '
            darcsec = '-1.00  '
            orient = '-1.00\n'
            if epoch_num < 10:
                vis = '0' + str(month) + '  '
                exp = '0' + str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + '0' + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
            else:
                vis = str(month) + '  '
                exp = str(placehold) + '  '
                epoch.write(file_name + pid + vis + exp + filter1 + str(epoch_num) 
                            + '  ' + mjd + exptime + darcsec + orient)
    
    epoch_num +=1
    epoch.write('\n')

#epoch.write(file_name+'\n')

epoch.close()
```
